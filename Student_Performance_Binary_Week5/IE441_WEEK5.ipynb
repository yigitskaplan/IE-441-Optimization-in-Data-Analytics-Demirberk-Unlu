```python
import pandas as pd
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import numpy as np
# libraries imported
```

```python
studentperf = pd.read_csv("Student_Performance_binary.csv")
```

```python
studentperf.head()
```

```python
# Performance index binary, Kaldı | Geçti
```

```python
studentperf.drop(studentperf.columns[0], axis = 1, inplace = True)
```

```python
# 1. satırı veriden çıkartıyoruz. Saymaya 0 dan başlandığı için studentperf.columns[0] yazdık.
```

```python
studentperf.head()
```

```python
Y = studentperf['Performance Index']
X = studentperf.drop('Performance Index', axis = 1)
```

```python
X.head()
```

```python
# Extracurricular Activities, Yes:1 No:0 yapacağız
```

```python
X['Extracurricular Activities'] = X['Extracurricular Activities'].replace({'No':0, 'Yes': 1})
```

```python
X.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8, random_state = 42)
```

```python
x_train_sm = sm.add_constant(x_train)
```

```python
logml = sm.Logit(y_train, x_train_sm).fit()
```

```python
print(logml.summary())
```

```python
# logaritmik olarak değişiklerini veriyor
```

```python
print(np.exp(logml.params))
```

```python
# x lerin odd oranına değişimini veriyor
```

```python
print(logml.get_margeff().summary())
```

```python
# dy/dx değişim oranını veriyor, örn. Hours Studied değişkeni bir atrtırılrsa, başarılı olma olasılığı 0.0463 artıyor.
```

```python
y_hat = logml.predict(x_train_sm)
```

```python
y_hat
```

```python
# 1 sınıfında(başarılı) olma olasılıkları
```

```python
y_hat_b = (y_hat>0.5).astype(int)
```

```python
y_hat_b
```

```python
# 0.5 üstünde kalanlara 1, altında kalanlara 0 dedik
```

```python
report = classification_report(y_train,y_hat_b)
```

```python
print(report)
```

```python
print(confusion_matrix(y_train,y_hat_b))
```

```python
x_test_sm = sm.add_constant(x_test)
```

```python
logml_test = sm.Logit(y_test, x_test_sm).fit()
```

```python
print(logml_test.summary())
```

```python
print(np.exp(logml_test.params))
```

```python
print(logml_test.get_margeff().summary())
```

```python
y_hat_test = logml_test.predict(x_test_sm)
```

```python
y_hat_test_b = (y_hat_test>0.5).astype(int)
```

```python
report = classification_report(y_test,y_hat_test_b)
```

```python
print(report)
```
